Implemente uma Lista Encadeada Simples em Python, conforme discutido em aula.

A implementação deve utilizar classes próprias para representar:

Node → representa cada nó da lista

LinkedList → representa a lista encadeada

Cada nó deve armazenar:

um valor (data)

uma referência para o próximo nó (next)

```python
#Node → representa cada nó da lista

class Node:
    def __init__(self, data=None): #Construtor do nó
        self.data = data #Dado que o nó irá armazenar (Vem através do parâmetro no construtor)
        self.next = None #Ponteiro para o próximo nó (Inicialmente None)
        
    def __str__(self): #Método para representar o nó como string (útil para impressão)
        return f"Node(data={self.data}, next={self.next})"
        
no1 = Node(2)
no2 = Node(4)
no3 = Node(8)
no4 = Node(16)
no1.next = no2
no2.next = no3
no3.next = no4
print(no1)
```

```python
#LinkedList → representa a lista encadeada

class LinkedList:
    def __init__(self): #Construtor de listas encadeadas
        self.head = None #Cabeça da lista
        self._size = 0 #Tamanho da lista

    def insert_beginning(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node
        self._size += 1
    
    def insert_end(self, value):
        if not self.head:
            self.head = Node(value)
        else:
            pointer = self.head
            while pointer.next: #Enquanto tiver um next, vai seguir caminhando
                pointer = pointer.next
            pointer.next = Node(value) #Quando não tiver mais next, insere o novo nó (ao final)
    
    def remove(self, value):
        pointer = self.head
        previous_node = None
        counter = 0
        while pointer:
            if pointer.data == value:
                if previous_node:
                    previous_node.next = pointer.next
                else:
                    self.head = pointer.next
                self._size -= 1
                print(f'Pronto: {value} encontrado e removido junto com {counter+1}º elemento desta lista encadeada.')
                return
            else:
                previous_node = pointer
                pointer = pointer.next
            counter += 1
        print(f'Atenção: Não há elemento com o valor:{value} nesta lista encadeada.')
        return
    
    def search(self, value):
        pointer = self.head
        counter = 0
        while pointer:
            if pointer.data == value:
                print(f'Pronto: {value} encontrado no {counter+1}º elemento desta lista encadeada.')
                return
            else:
                pointer = pointer.next
            counter += 1
        print(f'Atenção: Não há elemento com o valor:{value} nesta lista encadeada.')
        return
        
    def print_list(self):
        pointer = self.head
        counter = 1
        if pointer:
            for i in range(self._size):
                print(f'{counter}º - {pointer.data}')
                pointer = pointer.next
                counter +=1
        return
    
    def size(self):
        print(f'Essa lista possui {self._size} itens')
        
    def is_empty(self):
        if self.head:
            print('Essa lista possui ao menos um elemento.')
        else:
            print('Essa lista não possui elementos.')
    
lista2 = LinkedList()
lista2.insert_beginning(2)
lista2.insert_beginning(4)
lista2.remove(2)
lista2.insert_beginning(8)
lista2.insert_beginning(16)
lista2.insert_beginning(32)
lista2.insert_beginning(64)
lista2.insert_beginning(128)
lista2.insert_beginning(256)
lista2.remove(32)
lista2.search(8)
lista2.print_list()
lista2.size()
lista3 = LinkedList()

lista2.is_empty()
lista3.is_empty()
```
