Implemente uma Lista Encadeada Simples em Python, conforme discutido em aula.

A implementação deve utilizar classes próprias para representar:

Node → representa cada nó da lista

LinkedList → representa a lista encadeada

Cada nó deve armazenar:

um valor (data)

uma referência para o próximo nó (next)

```python
#Node → representa cada nó da lista

class Node:
    def __init__(self, data=None): #Construtor do nó
        self.data = data #Dado que o nó irá armazenar (Vem através do parâmetro no construtor)
        self.next = None #Ponteiro para o próximo nó (Inicialmente None)
        
    def __str__(self): #Método para representar o nó como string (útil para impressão)
        return f"Node(data={self.data}, next={self.next})"
        
no1 = Node(2)
no2 = Node(4)
no3 = Node(8)
no4 = Node(16)
no1.next = no2
no2.next = no3
no3.next = no4
print(no1)
```
